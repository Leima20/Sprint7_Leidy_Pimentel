**1-Funciones de Entrada y Validación**

*solicitar_texto(mensaje)*
* Inicia un bucle infinito para garantizar la entrada.
* Solicita la entrada por teclado y elimina espacios innecesarios.
* Valida si el campo tiene contenido; si es así, lo retorna.
* Si está vacío, muestra un error y repite.

```python
def solicitar_texto(mensaje):
    while True:
        valor = input(mensaje).strip()
        if valor: return valor
        print("Error: El campo no puede estar vacío.")
```

**solicitar_numero(mensaje)**
* Solicita la entrada por teclado.
* Intenta convertir la entrada a un número entero.
* Si tiene éxito, lo retorna.
* Si falla (ValueError), retorna -1 como código de error.

```python
def solicitar_numero(mensaje):
    valor = input(mensaje).strip()
    try:
        return int(valor)
    except ValueError:
        return -1
```

2. Lógica de Negocio y Clasificacióncalcular_puntuacion(experiencia, catalan, excel, titulacion)Verifica si los valores numéricos están fuera del rango permitido (0-3 para competencias, >=0 para experiencia).Si hay error, retorna -1.0.Si son válidos, aplica la ponderación: $(exp * 1.5) + cat + exc + tit$.Retorna el resultado final.

```python
def calcular_puntuacion(experiencia, catalan, excel, titulacion):
    if (catalan < 0 or catalan > 3 or
        excel < 0 or excel > 3 or
        titulacion < 0 or titulacion > 3 or
        experiencia < 0):
        return -1.0
    return (experiencia * 1.5) + catalan + excel + titulacion
```

obtener_prioridad(puntuacion)
Si la puntuación es -1.0, retorna un error y peso 0.

Evalúa los rangos (>=12, >=8) para asignar la categoría.

Retorna la etiqueta textual y un valor numérico para facilitar el ordenamiento.

```python
def obtener_prioridad(puntuacion):
    if puntuacion == -1.0: return "ERROR: Datos Inválidos", 0
    if puntuacion >= 12: return "Prioridad Alta", 3
    if puntuacion >= 8: return "Candidato adecuado", 2
    return "No prioritario", 1
```

Funciones Modulares



Pseudocódigo: obtener_segmento_y_factor
* Recorre cada cliente del diccionario.
* Busca en 'comportamientos' usando el código del cliente.
* Asigna el segmento y factor correspondiente al cliente.

```python
"""
    Actualiza el diccionario de clientes añadiendo el segmento,
    el mensaje y el factor de comportamiento basándose en el
    código de comportamiento de cada cliente.
    """

def obtener_segmento_y_factor(diccionario):
    for cliente in diccionario:
        codigo = diccionario[cliente]["codigo_comportamiento"]
        diccionario[cliente]["segmento"] = comportamientos[codigo]["segmento"]
        diccionario[cliente]["factor_comportamiento"] = comportamientos[codigo]["factor"]
```

Pseudocódigo: obtener_factor_edad
* Recorre cada cliente del diccionario.
* Busca en 'edades' usando el grupo_edad del cliente.
* Asigna el factor de edad al cliente.

```python
"""
    Recorre el diccionario de clientes y añade el factor de riesgo
    basado en el grupo de edad de cada uno.
    """

def obtener_factor_edad(diccionario):
    for cliente in diccionario:
        grup = diccionario[cliente]["grupo_edad"]
        diccionario[cliente]["factor_edad"] = edades[grup]
```

Pseudocódigo: calcular_riesgo
* Recorre cada cliente del diccionario.
* Multiplica el factor_comportamiento por el factor_edad.
* Almacena el resultado en 'riesgo_total'.

```python
"""
    Realiza el cálculo del riesgo multiplicando el factor de comportamiento
    por el de edad, redondeando el resultado a 3 decimales.
    """

def calcular_riesgo(diccionario):
    for cliente in diccionario:
        riesgo = diccionario[cliente]["factor_comportamiento"] * diccionario[cliente]["factor_edad"]
        diccionario[cliente]["riesgo_total"] = round(riesgo, 3)
```

**Función: processar_clients**

Pseudocódigo: processar_clients
* Crea un nuevo diccionario organizado por id_client.
* Llama secuencialmente a las funciones de cálculo (obtener datos, factores, riesgo).
* Retorna el diccionario enriquecido.

```python
def processar_clients(clientes_lista):
    diccionario = {
        c["id_client"]: {k: v for k, v in c.items() if k != "id_client"}
        for c in clientes_lista
    }
    obtener_segmento_y_factor(diccionario)
    obtener_factor_edad(diccionario)
    calcular_riesgo(diccionario)
    return diccionario
```

PSEUDOCÓDIGO: mostrar_diccionari
1. Imprime el encabezado y el diccionario procesado.

```python
def mostrar_diccionari(diccionario):
    print("Diccionario de clientes con sus riesgos:\n")
    print(diccionario)
```

Pseudocódigo: mostrar_respuestas
* Crear una lista de clientes a partir de los valores del diccionario.
* Ordenar por riesgo de mayor a menor y mostrar los 3 primeros.
* Filtrar clientes con código 4 para analizar cómo la edad afecta al riesgo.
* Filtrar clientes con riesgo < 0.2 (poco interés).
* Seleccionar el cliente de mayor riesgo (más rentable a recuperar).

```python
def mostrar_respuestas(diccionario):
    clientes_lista = list(diccionario.values())

    print("\n1. Clientes a contactar primero (Retargeting):")
    top_riesgo = sorted(clientes_lista, key=lambda x: x['riesgo_total'], reverse=True)
    for c in top_riesgo[:3]:
        print(f"- {c['nombre']} (Riesgo: {c['riesgo_total']:.2f})")

    print("\n2. Diferencias en clientes con código 4 (Impacto edad):")
    analisis_c4 = [c for c in clientes_lista if c['codigo_comportamiento'] == 4]
    for c in analisis_c4:
        print(f"- {c['nombre']} (Edad: {c['grupo_edad']}, Riesgo: {c['riesgo_total']:.2f})")

    print("\n3. Clientes con bajo interés (Riesgo < 0.2):")
    poc_interes = [c['nombre'] for c in clientes_lista if c['riesgo_total'] < 0.2]
    print(", ".join(poc_interes))

    print("\n4. Cliente más rentable a recuperar:")
    mas_rentable = max(clientes_lista, key=lambda x: x['riesgo_total'])
    print(f"- {mas_rentable['nombre']} (Riesgo: {mas_rentable['riesgo_total']:.2f})")
```

Función: main_marketing

* PSEUDOCÓDIGO: main_marketing
* Orquesta todo el programa llamando a procesar, mostrar diccionario y mostrar respuestas.

```python
"""
    Función orquestadora que gestiona el flujo de procesamiento de los clientes
    y dispara la generación del reporte final.
    """

def main_marketing():
    diccionario = processar_clients(clientes)
    mostrar_diccionari(diccionario)
    mostrar_respuestas(diccionario)
```

Gestió de risc i priorització de pacients
Treballes com a analista de dades en un centre mèdic especialitzat. Diversos pacients han enviat els seus símptomes a través d'un formulari online, però les dades arriben en un format desordenat i poc estructurat, i l'equip mèdic necessita que els organitzis per poder prioritzar l'ordre d'atenció segons la gravetat.

Objectius
L'objectiu és transformar aquestes dades en informació útil per als metges. A partir dels registres en brut, has de:

* Convertir cada string en una estructura neta (diccionari o similar).

* Separar i normalitzar la llista de símptomes:
minúscules
eliminar espais
dividir correctament per comes
* Assignar un risc a cada símptoma utilitzant el diccionari nivells_risc.

* Calcular el risc total de cada pacient sumant els valors.

* Ordenar els pacients de major a menor risc per determinar la prioritat d'atenció.

Resultat esperat
Un DataFrame o un diccionari amb:

nom
edat
símptomes normalitzats
risc_total
I un apartat final amb els pacients ordenats per prioritat (més risc → primer).

```python
# Definimos los riesgos fuera de las funciones
NIVELES_RIESGO = {
    "fiebre alta": 3,
    "dificultad respirar": 5,
    "dolor pecho": 4,
    "tos": 1,
    "fatiga": 1,
    "dolor cabeza": 1
}
```

Funciones Modulares

Función: normalizar_síntomas

Algoritmo:

* Recibir el string de síntomas en bruto.

* Separar el string mediante el delimitador de coma.

* Para cada síntoma: quitar espacios, pasar a minúsculas y reemplazar guiones bajos por espacios.

* Retornar una lista limpia de síntomas.

```python
def normalizar_sintomas(lista_sintomas):
    """Limpia y normaliza la lista de síntomas en formato texto."""
    return [s.strip().lower().replace("_", " ") for s in lista_sintomas.split(",")]
```

Función: calcular_riesgo_total

Algoritmo:

Recibir la lista de síntomas ya normalizados.

Recorrer la lista y sumar el valor de riesgo buscando cada síntoma en NIVELES_RIESGO.

Si un síntoma no existe en el diccionario, sumar 0.

Retornar el riesgo total calculado.

```python
def calcular_riesgo_total(lista_síntomas, diccionario_riesgos):
    """Suma los valores de riesgo de una lista de síntomas."""
    return sum([diccionario_riesgos.get(s, 0) for s in lista_síntomas])
```

Función: procesar_pacientes

Algoritmo:

Inicializar una lista vacía para almacenar los pacientes procesados.

Recorrer la lista de pacientes original.

Dividir el string por el delimitador "|" para obtener nombre, edad y síntomas.

Llamar a normalizar_simptomas y calcular_riesgo_total.

Crear un diccionario con los datos estructurados.

Retornar la lista final.

```python
def procesar_pacientes(pacientes):
    """Convierte strings crudos en una lista de diccionarios estructurados."""
    resultados = []
    for p in pacientes:
        datos = p.split("|")
        sintomas_limpios = normalizar_sintomas(datos[2])

        resultados.append({
            "nombre": datos[0],
            "edad": int(datos[1]),
            "sintomas_normalizados": sintomas_limpios,
            "riesgo_total": calcular_riesgo_total(sintomas_limpios, NIVELES_RIESGO)
        })
    return resultados
```

Función: main_triage

Algoritmo:

Recibir la lista cruda de pacientes.

Llamar a procesar_pacientes para obtener la estructura limpia.

Ordenar la lista por riesgo_total de mayor a menor utilizando sorted().

Retornar la lista ordenada.

```python
def main_triage(pacientes):
    """Orquesta el procesamiento y la priorización de pacientes."""
    datos_limpios = procesar_pacientes(pacientes)
    return sorted(datos_limpios, key=lambda x: x['riesgo_total'], reverse=True)
```

Treballes com a analista de dades en una companyia asseguradora.
L'empresa vol calcular el risc dels clients, el preu estimat de la pòlissa i la probabilitat de frau, utilitzant un codi de comportament que ha estat generat per un model intern.

El preu final s'ha de calcular amb la fórmula:

preu_final = preu_base * (1 + edat/100)

Objectius
» Classificar cada client utilitzant nivells_seguro (sense if/elif).

» Afegir:
La categoria
El preu base
La probabilitat de frau
» Calcular el preu final de la pòlissa amb la fórmula donada.

» Crear un DataFrame o diccionari amb la informació final.

» Ordenar els clients:

Per preu final.

Resultat esperat
Un DataFrame o un diccionari amb:

* Nom
* Edat
* Categoria
* Preu_base
* Preu_final
* Probabilitat_frau
* Un apartat d'anàlisi on s'indiqui:
* Quin client té més probabilitat de frau
* Quin té el preu final més alt
* Qui seria prioritari de revisar manualment

Inicialización de Datos

Definir una lista de diccionarios llamada clients que contenga: nombre, edad y código de comportamiento de cada individuo.

Definir un diccionario llamado nivells_seguro que actúe como base de datos, donde la clave sea el código y el valor sea otro diccionario con: categoría, precio base y probabilidad de fraude.

```python
# Datos de entrada
clients = [
    {"nom": "Ana", "edat": 42, "codi": 0},
    {"nom": "Carlos", "edat": 30, "codi": 2},
    {"nom": "Isabel", "edat": 55, "codi": 3},
    {"nom": "Jorge", "edat": 40, "codi": 1},
    {"nom": "Marta", "edat": 28, "codi": 0}
]

nivells_seguro = {
    0: {"categoria": "Baix risc", "preu_base": 120, "fraude": 0.01},
    1: {"categoria": "Risc mitjà", "preu_base": 200, "fraude": 0.05},
    2: {"categoria": "Risc alt", "preu_base": 350, "fraude": 0.15},
    3: {"categoria": "Risc crític", "preu_base": 500, "fraude": 0.30}
}
```

Algorítmo: calcular_precio_final Recibir precio_base y edad.

Calcular: resultado = precio_base * (1 + edad / 100).

Retornar resultado.

```python
def calcular_precio_final(preu_base, edad):
    return preu_base * (1 + edad / 100)
```

Algoritmo de procesar_cliente:

Recibir cliente y el diccionario de niveles.

Obtener la información de riesgo usando cliente['codi'] como clave.

Crear un diccionario con: Nombre, Edad, Categoría, Precio_base, Probabilidad_fraude.

Añadir Precio_final llamando a la función calcular_precio_final.

Retornar el nuevo diccionario.

```python
def procesar_cliente(cliente, niveles):
    info = niveles[cliente['codi']]
    return {
        "Nombre": cliente['nom'],
        "Edad": cliente['edat'],
        "Categoría": info['categoria'],
        "Precio_base": info['preu_base'],
        "Probabilidad_fraude": info['fraude'],
        "Precio_final": calcular_precio_final(info['preu_base'], cliente['edat'])
    }
```

Algoritmo de gestionar_aseguradora:

Recibir lista_clientes y niveles.

Aplicar procesar_cliente a cada elemento de la lista.

Crear lista_por_fraude: ordenar los datos por Probabilidad_fraude (descendente).

Crear lista_por_precio: ordenar los datos por Precio_final (ascendente).

Retornar ambas listas.

```python
def gestionar_aseguradora(lista_clientes, niveles):
    datos = [procesar_cliente(c, niveles) for c in lista_clientes]

    por_fraude = sorted(datos, key=lambda x: x['Probabilidad_fraude'], reverse=True)
    por_precio = sorted(datos, key=lambda x: x['Precio_final'])

    return por_fraude, por_precio
```

Algorítmo de Análisis Final:

Ejecutar la función principal para obtener las listas ordenadas.

Imprimir los resultados en formato de tabla.

Identificar y mostrar el cliente con más fraude, el precio más alto y el prioritario de revisión.

```python
# Ejecución y Análisis
lista_fraude, lista_precio = gestionar_aseguradora(clients, nivells_seguro)

# Visualización de datos
print(f"{'Nombre':<10} | {'Categoría':<12} | {'Precio Final':<12} | {'Fraude':<6}")
for c in lista_fraude:
    print(f"{c['Nombre']:<10} | {c['Categoría']:<12} | {c['Precio_final']:<12.2f} | {c['Probabilidad_fraude']}")

print("\n--- APARTADO DE ANÁLISIS ---")
print(f"1. Cliente con más probabilidad de fraude: {lista_fraude[0]['Nombre']} ({lista_fraude[0]['Probabilidad_fraude']*100}%)")
print(f"2. Cliente con el precio final más alto: {lista_precio[-1]['Nombre']} ({lista_precio[-1]['Precio_final']:.2f}€)")
print(f"3. Prioritario de revisar manualment: {lista_fraude[0]['Nombre']} (por mayor riesgo de fraude)")
```
