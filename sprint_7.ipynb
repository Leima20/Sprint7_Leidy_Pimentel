FUNCIÓN pedir_entero(mensaje, minimo, maximo)

    REPETIR
        INTENTAR
            Mostrar mensaje
            Leer valor
            Convertir valor a entero

            SI existe un mínimo Y valor < mínimo ENTONCES
                Mostrar mensaje de error
                Continuar repitiendo
            FIN SI

            SI existe un máximo Y valor > máximo ENTONCES
                Mostrar mensaje de error
                Continuar repitiendo
            FIN SI

            Devolver valor

        SI ocurre un error de conversión ENTONCES
            Mostrar mensaje indicando que debe introducir un número entero
        FIN INTENTAR

    HASTA que el valor sea válido

Fin Función

```python
def pedir_entero(mensaje, minimo=None, maximo=None):
    while True:
        try:
            valor = int(input(mensaje))

            if minimo is not None and valor < minimo:
                print(f"Error: El valor debe ser como mínimo {minimo}.\n")
                continue

            if maximo is not None and valor > maximo:
                print(f"Error: El valor debe ser como máximo {maximo}.\n")
                continue

            return valor

        except ValueError:
            print("Error: Debes introducir un número entero.\n")
```

FIN FUNCIÓN

```python
def obtenir_dades():
    nombre = input("Introduce el nombre y apellidos:\n")
    telefono = input("Introduce el teléfono:\n")

    años = pedir_entero("Introduce los años de experiencia:\n", minimo=0)

    nivel_catalan = pedir_entero(
        "Introduce el nivel de catalán (0-3):\n"
        "0 = sense coneixement\n"
        "1 = fins a nivell B1\n"
        "2 = nivell B2\n"
        "3 = C1 o superior\n",
        minimo=0,
        maximo=3,
    )

    nivel_excel = pedir_entero(
        "Introduce el nivel de Excel (0-3):\n"
        "0 = nivel básico\n"
        "1 = uso de funciones\n"
        "2 = tablas dinámicas\n"
        "3 = Visual Basic y Macros\n",
        minimo=0,
        maximo=3,
    )

    nivel_academico = pedir_entero(
        "Introduce el nivel académico (0-3):\n"
        "0 = ESO\n"
        "1 = FP\n"
        "2 = Universitat\n"
        "3 = Màster o superior\n",
        minimo=0,
        maximo=3,
    )

    return [
        nombre,
        telefono,
        años,
        nivel_catalan,
        nivel_excel,
        nivel_academico,
    ]
```

**FUNCIÓN validar_dades(datos_candidato)**

  * SI la lista no contiene exactamente 6 elementos ENTONCES
        
    * Devolver -1

    FIN SI

  * SI el nombre no contiene al menos dos palabras ENTONCES
      
    * Devolver 1
    
    FIN SI

  * SI el teléfono no tiene exactamente 9 cifras

    O contiene caracteres no numéricos ENTONCES
      
     * Devolver 2
    
    FIN SI

  * SI los años de experiencia son negativos ENTONCES

    * Devolver

    FIN SI

  * PARA cada dato desde nivel de catalán hasta nivel académico

      SI el valor no está entre 0 y 3 ENTONCES

    * Devolver el código de error correspondiente

   FIN SI

  FIN PARA

   * Devolver 0

FIN FUNCIÓN

```python
def validar_dades(datos_candidato: list):
  if len(datos_candidato) != 6:
    return -1
  if len(datos_candidato[0].split()) < 2:
    return 1
  if len(datos_candidato[1]) != 9 or not datos_candidato[1].isdigit():
    return 2
  if datos_candidato[2] < 0:
    return 3
  for i in range(3, 6):
      if not 0 <= datos_candidato[i] <= 3:
          return i + 1
  return 0
```

FUNCIÓN calcular_puntuacio(datos_candidato)

    SI la lista no contiene exactamente 6 elementos ENTONCES
        Devolver -1
    FIN SI

    puntuación ←
        (años de experiencia × 1,5)
        + nivel de catalán
        + nivel de Excel
        + nivel académico

    Devolver puntuación

FIN FUNCIÓN

```python
def calcular_puntuacio(datos_candidato: list):
  if len(datos_candidato) != 6:
    return -1
  return datos_candidato[2]*1.5 + datos_candidato[3] + datos_candidato[4] + datos_candidato[5]
```

FUNCIÓN classificar_candidat(puntuacion)

    SI puntuación es mayor o igual que 12 ENTONCES
        Devolver prioridad 3
        Devolver "Prioridad alta"

    SI NO, SI puntuación es mayor o igual que 8 ENTONCES
        Devolver prioridad 2
        Devolver "Candidat adequat"

    EN CASO CONTRARIO
        Devolver prioridad 1
        Devolver "No prioritari"

FIN FUNCIÓN

```python
def classificar_candidat(puntuacion):
  if puntuacion >= 12:
    return 3, "Prioridad alta"
  elif puntuacion >= 8:
    return 2 , "Candidat adequat"
  else:
    return 1, "No prioritari"
```

FUNCIÓN mostrar_individual(datos_candidato)

    Mostrar "Datos del candidato"

```python
def mostrar_individual(datos_candidato):
  print("Datos del candidato:\n")
  print(f"Nombre y apellidos: {datos_candidato[0]}")
  print(f"Teléfono: {datos_candidato[1]}")
  print(f"Años de experiencia: {datos_candidato[2]}")
  print(f"Nivel de catalán: {datos_candidato[3]}")
  print(f"Nivel de Excel: {datos_candidato[4]}")
  print(f"Nivel académico: {datos_candidato[5]}")
  print(f"Puntuación: {datos_candidato[6]}")
  print(f"Clasificación: {datos_candidato[8]}")
```

FUNCIÓN mostrar_informe(candidatos)

    Mostrar el encabezado del informe

    Ordenar los candidatos
        Primero por prioridad
        Después por puntuación
        En orden descendente

    PARA cada candidato ordenado

        Mostrar:
            clasificación
            nombre
            puntuación

    FIN PARA

FIN FUNCIÓN

```python
def mostrar_informe(candidatos: list):
  print("\n" + "="*75)
  print("INFORME FINAL DE SELECCIÓN - DEPARTAMENTO DE RRHH")
  print("="*75)
  print(f"{'Prioridad':<20} | {'Nombre':<30} | {'Puntos'}")
  print("="*75)
  for c in sorted(candidatos, key=lambda x: (x[7], x[6]), reverse=True):
    print(f"{c[8]:<20} | {c[0]:<30} | {c[6]}")
```

FUNCIÓN main_seleccio

    Crear lista vacía de candidatos

    contador ← 0

    MIENTRAS contador sea menor que 10

        Mostrar mensaje para introducir un candidato

        datos_candidato ← obtenir_dades()

        resultado ← validar_dades(datos_candidato)

        SEGÚN resultado HACER

            -1:
                Mostrar que la lista de datos no tiene 6 elementos
            1:
                Mostrar error en el nombre
            2:
                Mostrar error en el teléfono
            3:
                Mostrar error en los años de experiencia
            4:
                Mostrar error en el nivel de catalán
            5:
                Mostrar error en el nivel de Excel
            6:
                Mostrar error en el nivel académico
            0:
                Mostrar que los datos son correctos

                puntuación ← calcular_puntuacio(datos_candidato)

                Añadir puntuación a la lista

                prioridad, clasificación ←
                    classificar_candidat(puntuación)

                Añadir prioridad a la lista

                Añadir clasificación a la lista

                Mostrar los datos del candidato

                Añadir candidato a la lista de candidatos

                Incrementar contador en 1

        FIN SEGÚN

    FIN MIENTRAS

    Mostrar el informe final

FIN FUNCIÓN

```python
def main_seleccio():
  candidatos = []
  contador = 0
  while contador < 10:
    print(f"\nIntroduce los datos del candidato.")
    datos_candidato = obtenir_dades()
    match validar_dades(datos_candidato):
      case -1: print("Los datos del cantidato no son 6.")
      case 1: print("Hay un error en el nombre del candidato.")
      case 2: print("Hay un error en el teléfono del candidato.")
      case 3: print("Hay un error en los años de experiencia del candidato.")
      case 4: print("Hay un error en el nivel de Catalán del candidato.")
      case 5: print("Hay un error en el nivel de Excel del candidato.")
      case 6: print("Hay un error en el nivel de académico del candidato.")
      case 0:
        print("Los datos del cantidato son correctos.\n")
        datos_candidato.append(calcular_puntuacio(datos_candidato))
        prioridad, clasificacion = classificar_candidat(datos_candidato[6])
        datos_candidato.append(prioridad)
        datos_candidato.append(clasificacion)
        mostrar_individual(datos_candidato)
        candidatos.append(datos_candidato)
        contador += 1
  mostrar_informe(candidatos)
```

**Algoritmo General:**

Funciones Modulares



Pseudocódigo: obtener_segmento_y_factor
* Recorre cada cliente del diccionario.
* Busca en 'comportamientos' usando el código del cliente.
* Asigna el segmento y factor correspondiente al cliente.

```python
def obtener_segmento_y_factor(diccionario: {}):
    for cliente in diccionario:
        diccionario[cliente]["segmento"] = comportamientos[diccionario[cliente]["codigo_comportamiento"]]["segmento"]
        diccionario[cliente]["factor_comportamiento"] = comportamientos[diccionario[cliente]["codigo_comportamiento"]]["factor"]
```

Pseudocódigo: obtener_factor_edad
* Recorre cada cliente del diccionario.
* Busca en 'edades' usando el grupo_edad del cliente.
* Asigna el factor de edad al cliente.

```python
def obtener_factor_edad(diccionario: {}):
    for cliente in diccionario:
        diccionario[cliente]["factor_edad"] = edades[diccionario[cliente]["grupo_edad"]]
```

Pseudocódigo: calcular_riesgo
* Recorre cada cliente del diccionario.
* Multiplica el factor_comportamiento por el factor_edad.
* Almacena el resultado en 'riesgo_total'.

```python
def calcular_riesgo(diccionario: {}):
    for cliente in diccionario:
        diccionario[cliente]["riesgo_total"] = round(diccionario[cliente]["factor_comportamiento"] * diccionario[cliente]["factor_edad"],3)
```

**Función: processar_clients**

Pseudocódigo: processar_clients
* Crea un nuevo diccionario organizado por id_client.
* Llama secuencialmente a las funciones de cálculo (obtener datos, factores, riesgo).
* Retorna el diccionario enriquecido.

```python
def processar_clients(clientes_lista):
    diccionario = {
        c["id_client"]: {k: v for k, v in c.items() if k != "id_client"}
        for c in clientes_lista
    }
    obtener_segmento_y_factor(diccionario)
    obtener_factor_edad(diccionario)
    calcular_riesgo(diccionario)
    return diccionario
```

PSEUDOCÓDIGO: mostrar_diccionari
1. Imprime el encabezado y el diccionario procesado.

```python
def mostrar_diccionari(diccionario):
    print("=" * 170)
    print("Diccionario de clientes con sus riesgos:\n")
    print("=" * 170)

    print(
        f"{'Nombre':<20} | {'Edad':<8} | {'Grupo Edad':<15} | "
        f"{'Código Comportamiento':<25} | {'Segmento':<20} | "
        f"{'Factor comportamiento':<22} | {'Factor edad':<15} | "
        f"{'Riesgo total':<15}"
    )

    print("=" * 170)

    for datos in diccionario.values():
        print(
            f"{datos['nombre']:<20} | "
            f"{datos['edad']:<8} | "
            f"{datos['grupo_edad']:<15} | "
            f"{datos['codigo_comportamiento']:<25} | "
            f"{datos['segmento']:<20} | "
            f"{datos['factor_comportamiento']:<22} | "
            f"{datos['factor_edad']:<15} | "
            f"{datos['riesgo_total']:<15}"
        )
```

Pseudocódigo: mostrar_respuestas
* Crear una lista de clientes a partir de los valores del diccionario.
* Ordenar por riesgo de mayor a menor y mostrar los 3 primeros.
* Filtrar clientes con código 4 para analizar cómo la edad afecta al riesgo.
* Filtrar clientes con riesgo < 0.2 (poco interés).
* Seleccionar el cliente de mayor riesgo (más rentable a recuperar).

```python
def mostrar_respuestas(diccionario):
    clientes_lista = list(diccionario.values())

    print("\n1. Clientes a contactar primero (Retargeting):")
    top_riesgo = sorted(clientes_lista, key=lambda x: x['riesgo_total'], reverse=True)
    for c in top_riesgo[:3]:
        print(f"- {c['nombre']} (Riesgo: {c['riesgo_total']:.2f})")

    print("\n2. Los clientes con mismo codigo de comportamiento exhiben diferentes niveles de riesgo total en base a sus distintos factores de edad. Ejemplo para comportamiento=4:")
    analisis_c4 = [c for c in clientes_lista if c['codigo_comportamiento'] == 4]
    for c in analisis_c4:
        print(f"- {c['nombre']} (Edad: {c['grupo_edad']}, Riesgo: {c['riesgo_total']:.2f})")

    print("\n3. Clientes con bajo interés (Riesgo < 0.2):")
    poc_interes = [c['nombre'] for c in clientes_lista if c['riesgo_total'] < 0.2]
    print(", ".join(poc_interes))

    print("\n4. Cliente más rentable a recuperar:")
    mas_rentable = max(clientes_lista, key=lambda x: x['riesgo_total'])
    print(f"- {mas_rentable['nombre']} (Riesgo: {mas_rentable['riesgo_total']:.2f})")
```

Función: main_marketing

```python
def main_marketing():
    diccionario = processar_clients(clientes)
    mostrar_diccionari(diccionario)
    mostrar_respuestas(diccionario)
```

Diversos pacients han enviat els seus símptomes a través d'un formulari online, però les dades arriben en un format desordenat i poc estructurat, i l'equip mèdic necessita que els organitzis per poder prioritzar l'ordre d'atenció segons la gravetat.

Objectius
L'objectiu és transformar aquestes dades en informació útil per als metges. A partir dels registres en brut, has de:

* Convertir cada string en una estructura neta (diccionari o similar).

* Separar i normalitzar la llista de símptomes:
minúscules
eliminar espais
dividir correctament per comes
* Assignar un risc a cada símptoma utilitzant el diccionari nivells_risc.

* Calcular el risc total de cada pacient sumant els valors.

* Ordenar els pacients de major a menor risc per determinar la prioritat d'atenció.

Resultat esperat
Un DataFrame o un diccionari amb:

nom
edat
símptomes normalitzats
risc_total
I un apartat final amb els pacients ordenats per prioritat (més risc → primer).

**Algoritmo General:**


Gestión de Riesgo y Priorización de Pacientes

Configuración: Definición de los niveles de riesgo y estructura de datos inicial.

Entrada y Limpieza: Conversión de cadenas crudas a diccionarios estructurados, normalizando síntomas.

Lógica de Negocio: Cálculo de la puntuación total de riesgo sumando los valores asignados a cada síntoma.

Orquestación: Secuencia lógica que encadena la limpieza, el cálculo y la ordenación de los casos.

Salida: Generación de un informe estructurado que muestra a los pacientes ordenados por su nivel de riesgo.

**Desgranaje del Sistema Modular**

```python
# --- DATOS ---

pacientes = [
    "Maria|45|tos, Fiebre Alta, dolor_pecho",
    "Luis|33|dolor_cabeza,faTiga",
    "Sara|67|fiebre alta, dificultad RESPIRAR, tos",
    "Jordi|52|fatiga , tos",
    "Anna|29|fiebre alta,dolor_cabeza"
]

nivells_risc = {
    "fiebre_alta": 3,
    "dificultad_respirar": 5,
    "dolor_pecho": 4,
    "tos": 1,
    "fatiga": 1,
    "dolor_cabeza": 1
}
```

**Funciones Modulares**

 PSEUDOCÓDIGO: limpiar_datos
1. Recibe la lista de pacientes.
2. Bucle: divide por '|', convierte edad a entero y crea lista limpia de síntomas.
3. Retorna un diccionario con los datos estructurados.

```python
def limpiar_datos(pacientes: []):
    dict_pacientes = {}

    # Para cada paciente de la lista, hacer lo siguiente:
    for paciente in pacientes:
        # Separar los datos del paciente delimitados por el caracter "|".
        datos = paciente.split("|")

        #Se obtiene el nombre, la edad y la lista de síntomas.
        nombre = datos[0]
        edad = int(datos[1])
        sintomas_sin_normalizar = datos[2].split(',')

        sintomas_normalizados = []
        for sintoma in sintomas_sin_normalizar:
            sintomas_normalizados.append("_".join(sintoma.lower().split()))

        dict_pacientes[nombre] = {
            "edad" : edad,
            "sintomas" : sintomas_normalizados
        }

    return dict_pacientes
```

PSEUDOCÓDIGO: calcular_riesgo
1. Recibe el diccionario de pacientes.
2. Bucle: suma el valor de riesgo de cada síntoma consultando 'nivells_risc'.
3. Guarda el total en la clave 'riesgo'.

```python
def calcular_riesgo(dict_pacientes: {}):
    for paciente in dict_pacientes:
        riesgo = 0
        for sintoma in dict_pacientes[paciente]["sintomas"]:
            riesgo += nivells_risc[sintoma]
        dict_pacientes[paciente]["riesgo"] = riesgo
```

 PSEUDOCÓDIGO: ordenar_por_riesgo
 1. Recibe el diccionario con riesgos.
 2. Ordena los elementos basándose en la clave 'riesgo' de forma descendente.
 3. Retorna la lista ordenada.

```python
def ordenar_por_riesgo(dict_pacientes: {}):
    return sorted(
        dict_pacientes.items(),
        key=lambda elemento: elemento[1]["riesgo"],
        reverse=True
    )
```

PSEUDOCÓDIGO: main
1. Limpia los datos.
2. Calcula el riesgo.
3. Ordena los resultados y los imprime

```python
def main():
    dict_pacientes = limpiar_datos(pacientes)
    calcular_riesgo(dict_pacientes)
    print("Diccionario de pacientes:\n")
    print(dict_pacientes)

    print("\n")

    pacientes_ordernados = ordenar_por_riesgo(dict_pacientes)
    print("Lista de pacientes ordenados en base a su riesgo:\n")
    print(pacientes_ordernados)
```

```python
if __name__ == "__main__":
    main()
```

```python
# Datos de entrada

clientes = [
    {"nom": "Ana", "edat": 42, "codi": 0},
    {"nom": "Carlos", "edat": 30, "codi": 2},
    {"nom": "Isabel", "edat": 55, "codi": 3},
    {"nom": "Jorge", "edat": 40, "codi": 1},
    {"nom": "Marta", "edat": 28, "codi": 0}
]

nivells_seguro = {
    0: {"categoria": "Baix risc", "preu_base": 120, "fraude": 0.01},
    1: {"categoria": "Risc mitjà", "preu_base": 200, "fraude": 0.05},
    2: {"categoria": "Risc alt", "preu_base": 350, "fraude": 0.15},
    3: {"categoria": "Risc crític", "preu_base": 500, "fraude": 0.30}
}
```

**Funciones**

PSEUDOCÓDIGO: añadir_datos
1. Recibe la lista de clientes.
2. Bucle: Busca en 'nivells_seguro' por código y añade datos al diccionario del cliente.
3. No retorna nada (modifica la lista original).

```python
def añadir_datos():
    for cliente in clientes:
        cliente["categoria"] = nivells_seguro[cliente["codi"]]["categoria"]
        cliente["preu_base"] = nivells_seguro[cliente["codi"]]["preu_base"]
        cliente["fraude"] = nivells_seguro[cliente["codi"]]["fraude"]
```

PSEUDOCÓDIGO: calcular_precio_final
1. Recibe la lista de clientes.
2. Bucle: Calcula el precio aplicando la fórmula: preu_base * (1 + edat/100).
3. No retorna nada (añade la clave 'precio_final').

```python
def calcular_precio_final():
    for cliente in clientes:
        cliente["precio_final"] = cliente["preu_base"] * (1 + cliente["edat"]/100)
```

PSEUDOCÓDIGO: generar_diccionario
1. Recibe la lista de clientes.
2. Transforma la lista en un diccionario donde la clave es el nombre.
3. Retorna el nuevo diccionario.

```python
def generar_diccionario():
    return {
        cliente["nom"]: {k: v for k, v in cliente.items() if k != "nom"}
        for cliente in clientes
    }
```

PSEUDOCÓDIGO: generar_informe
1. Recibe el diccionario de clientes.
2. Imprime encabezados y formatea cada fila con alineación fija.

```python
def ordenar_por_fraude(diccionario: {}):
    return sorted(
        diccionario.items(),
        key=lambda elemento: elemento[1]["fraude"],
        reverse=True
    )
```

PSEUDOCÓDIGO: generar_analisis
1. Recibe el diccionario.
2. Calcula mediante funciones 'max' los valores críticos.
3. Imprime el resumen de auditoría.

```python
def ordenar_por_precio(diccionario: {}):
    return sorted(
        diccionario.items(),
        key=lambda elemento: elemento[1]["precio_final"],
        reverse=True
    )
```

PSEUDOCÓDIGO: main
1. Llama a las funciones de procesamiento.
2. Genera los informes y el análisis final.

```python
def main():
    añadir_datos()
    calcular_precio_final()

    diccionario = generar_diccionario()
    print("Diccionario de clientes:\n")
    print(diccionario)

    print("\n\n")

    lista_fraude = ordenar_por_fraude(diccionario)
    print("Lista de clientes ordernados por su probabilidad de fraude:\n")
    print(lista_fraude)

    lista_precio = ordenar_por_precio(diccionario)
    print("\n\n")
    print("Lista de clientes ordernados por su precio final:\n")
    print(lista_precio)
```

**Exercici 1: Generador de contrasenyes segures**

Ets analista de dades dins el departament de Ciberseguretat d'una empresa tecnològica. S'ha detectat que la majoria de treballadors continuen utilitzant contrasenyes molt dèbils com “Asdf1234”, dates d'aniversari o combinacions fàcils d'endevinar. Per aquest motiu, el responsable del departament t'ha demanat desenvolupar un generador de contrasenyes segures, totalment parametritzable i modular.

Requisits del generador
Has de crear una funció principal i totes les subfuncions necessàries per:

» Generar contrasenyes de longitud variable.
» Decidir si han d'incloure:
* Majúscules
* Minúscules
* Números
* Signes especials

» Assegurar-te que com a mínim apareix un caràcter de cada tipus seleccionat.

» Aleatoritzar completament la contrasenya.

» Retornar la contrasenya generada.

La funció principal del generador ha de tenir valors per defecte, exactament els següents:

majuscules=True
minuscules=True
numeros=True
signes=False
Aquests paràmetres són obligatoris i formen part del disseny del sistema.

```python
from numpy import random
import string

def crear_contraseña(longitud:int, mayusculas=True, minusculas=True, numeros=True, signos=False):

    if not any((mayusculas, minusculas, numeros, signos)):
        raise ValueError("Debes seleccionar al menos un tipo de caracteres.")

    if longitud < 1:
        raise ValueError("La contraseña debe de tener al menos un carácter")

    grupos = []

    if mayusculas:
        grupos.append(string.ascii_uppercase)

    if minusculas:
        grupos.append(string.ascii_lowercase)

    if numeros:
        grupos.append(string.digits)

    if signos:
        grupos.append(string.punctuation)

    contraseña = [random.choice(list(grupo)) for grupo in grupos]

    caracteres = "".join(grupos)

    # Completar la contraseña
    faltan = longitud - len(contraseña)

    contraseña.extend(random.choice(list(caracteres), size=faltan))

    # Mezclar el resultado
    random.shuffle(contraseña)

    return "".join(contraseña)
```

```python
def main():
    contraseña = crear_contraseña(12, mayusculas=True, minusculas=True, numeros=True, signos=True)
    print(f"Tu contraseña es: {contraseña}")
```

```python
if __name__ == "__main__":
    main()
```

**Exercici 2:**

 Processament automàtic de dades esportives
Ara formes part d'un equip d'analítica dins una organització esportiva que gestiona dades de la lliga catalana. La teva companya de feina vol automatitzar la lectura del seu arxiu històric de resultats i extreure les estadístiques més importants. T'ha demanat fer un petit sistema en Python que processi el fitxer "historic partits.txt"
Objectius
Mitjançant un codi modular, clar i documentat, el teu programa ha de calcular:

» 1.Quin equip és el més golejador.

» 2.Quin és el més golejat.

» 3.El nombre total de gols que ha fet cada equip.

» 4.La classificació final ordenada per punts on cada victòria suma 3 pts, cada empat 1 pts y cada derrota 0 pts.

En aquest nivell no es dona cap funció. Ets tu qui ha de dissenyar un flux complet. El codi ha de reflectir una bona organització interna.



```python
import pandas as pd

def lee_historico(nombre_fichero: str):

    df = pd.read_csv(
        nombre_fichero,
        sep="\t",
        header=None,
        names=["equipo_local", "resultado", "equipo_visitante"]
    )

    df[["goles_local", "goles_visitante"]] = (
        df["resultado"]
        .str.split("-", expand=True)
        .astype(int)
    )

    df = df.drop(columns="resultado")

    return df
```

```python
def calcula_goles_favor(df: pd.DataFrame):
    goles = pd.concat([
        df[["equipo_local", "goles_local"]]
            .rename(columns={
                "equipo_local": "equipo",
                "goles_local": "goles"
            }),

        df[["equipo_visitante", "goles_visitante"]]
            .rename(columns={
                "equipo_visitante": "equipo",
                "goles_visitante": "goles"
            })
    ])

    return goles.groupby("equipo")["goles"].sum().sort_values(ascending=False)
```

```python
def calcula_goles_contra(df: pd.DataFrame):
    goles = pd.concat([
        df[["equipo_local", "goles_visitante"]]
            .rename(columns={
                "equipo_local": "equipo",
                "goles_visitante": "goles"
            }),

        df[["equipo_visitante", "goles_local"]]
            .rename(columns={
                "equipo_visitante": "equipo",
                "goles_local": "goles"
            })
    ])

    return goles.groupby("equipo")["goles"].sum().sort_values(ascending=False)
```

```python
def calcula_classificacion(df: pd.DataFrame):

    #Partidos ganados.
    pg = (
        df[df["goles_local"] > df["goles_visitante"]]
        .groupby("equipo_local")
        .size().rename_axis("equipo")
        .add(
            df[df["goles_visitante"] > df["goles_local"]]
            .groupby("equipo_visitante")
            .size().rename_axis("equipo"),
            fill_value=0
        )
    )

    # partidos empatados.
    pe = (
        df[df["goles_local"] == df["goles_visitante"]]
        .groupby("equipo_local")
        .size().rename_axis("equipo")
        .add(
            df[df["goles_local"] == df["goles_visitante"]]
            .groupby("equipo_visitante")
            .size().rename_axis("equipo"),
            fill_value=0
        )
    )

    # Partidos perdidos.
    pp = (
        df[df["goles_local"] < df["goles_visitante"]]
        .groupby("equipo_local")
        .size().rename_axis("equipo")
        .add(
            df[df["goles_visitante"] < df["goles_local"]]
            .groupby("equipo_visitante")
            .size().rename_axis("equipo"),
            fill_value=0
        )
    )

    puntos = (pg * 3 + pe).sort_values(ascending=False)

    return pg, pe, pp, puntos
```

```python
def main():
    df = lee_historico("historic partits.txt")

    equipos = pd.unique(pd.concat([df["equipo_local"], df["equipo_visitante"]]))

    goles_favor = calcula_goles_favor(df)
    goles_contra = calcula_goles_contra(df)
    pg, pe, pp, puntos = calcula_classificacion(df)

    resultado = pd.concat(
        {
            "gf": goles_favor,
            "gc": goles_contra,
            "pg": pg,
            "pe": pe,
            "pp": pp,
            "puntos": puntos,
        },
        axis=1
    ).fillna(0).astype(int)

    resultado = resultado.sort_values("puntos", ascending=False)

    print("Dataframe resultante:")
    print(resultado)

    print(f"\nEquipo mas golejador: {goles_favor.idxmax()}")
    print(f"\nEquipo mas goleado: {goles_contra.idxmax()}")
    print("\nClasificación final:")

    tabla_puntos = puntos.reset_index()
    tabla_puntos.columns = ["equipo", "puntos"]
    tabla_puntos.insert(0, "posicion", range(1, len(tabla_puntos) + 1))

    print(tabla_puntos.to_string(index=False))

if __name__ == "__main__":
  main()
```
