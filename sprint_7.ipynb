**1-Funciones de Entrada y Validación**

*solicitar_texto(mensaje)*
* Inicia un bucle infinito para garantizar la entrada.
* Solicita la entrada por teclado y elimina espacios innecesarios.
* Valida si el campo tiene contenido; si es así, lo retorna.
* Si está vacío, muestra un error y repite.

```python
def solicitar_texto(mensaje):
    while True:
        valor = input(mensaje).strip()
        if valor: return valor
        print("Error: El campo no puede estar vacío.")
```

**solicitar_numero(mensaje)**
* Solicita la entrada por teclado.
* Intenta convertir la entrada a un número entero.
* Si tiene éxito, lo retorna.
* Si falla (ValueError), retorna -1 como código de error.

```python
def solicitar_numero(mensaje):
    valor = input(mensaje).strip()
    try:
        return int(valor)
    except ValueError:
        return -1
```

2. Lógica de Negocio y Clasificacióncalcular_puntuacion(experiencia, catalan, excel, titulacion)Verifica si los valores numéricos están fuera del rango permitido (0-3 para competencias, >=0 para experiencia).Si hay error, retorna -1.0.Si son válidos, aplica la ponderación: $(exp * 1.5) + cat + exc + tit$.Retorna el resultado final.

```python
def calcular_puntuacion(experiencia, catalan, excel, titulacion):
    if (catalan < 0 or catalan > 3 or
        excel < 0 or excel > 3 or
        titulacion < 0 or titulacion > 3 or
        experiencia < 0):
        return -1.0
    return (experiencia * 1.5) + catalan + excel + titulacion
```

obtener_prioridad(puntuacion)
Si la puntuación es -1.0, retorna un error y peso 0.

Evalúa los rangos (>=12, >=8) para asignar la categoría.

Retorna la etiqueta textual y un valor numérico para facilitar el ordenamiento.

```python
def obtener_prioridad(puntuacion):
    if puntuacion == -1.0: return "ERROR: Datos Inválidos", 0
    if puntuacion >= 12: return "Prioridad Alta", 3
    if puntuacion >= 8: return "Candidato adecuado", 2
    return "No prioritario", 1
```

```python
def generar_informe(candidatos):
    print("\n" + "="*75)
    print("INFORME FINAL DE SELECCIÓN - DEPARTAMENTO DE RRHH")
    print("="*75)
    candidatos_ordenados = sorted(candidatos, key=lambda x: (x['peso_prioridad'], x['puntuacion']), reverse=True)
    for c in candidatos_ordenados:
        puntos_mostrar = "N/A" if c['puntuacion'] == -1.0 else c['puntuacion']
        print(f"{c['prioridad']:<20} | {c['nombre']:<15} | Puntos: {puntos_mostrar}")
```

```python
#Configuración de tablas de búsqueda

# --- DATOS ---
clientes = [
    {"id_client": 1, "nombre": "Laura Serra", "edad": 23, "grupo_edad": "18-25", "codigo_comportamiento": 4},
    {"id_client": 2, "nombre": "Marc Vidal", "edad": 35, "grupo_edad": "26-40", "codigo_comportamiento": 3},
    {"id_client": 3, "nombre": "Ana López", "edad": 29, "grupo_edad": "26-40", "codigo_comportamiento": 2},
    {"id_client": 4, "nombre": "Joan Riera", "edad": 19, "grupo_edad": "18-25", "codigo_comportamiento": 2},
    {"id_client": 5, "nombre": "Mónica Pérez", "edad": 47, "grupo_edad": "41-60", "codigo_comportamiento": 4},
    {"id_client": 6, "nombre": "Luis García", "edad": 52, "grupo_edad": "41-60", "codigo_comportamiento": 1},
    {"id_client": 7, "nombre": "Pilar Sánchez", "edad": 61, "grupo_edad": "60+", "codigo_comportamiento": 3},
    {"id_client": 8, "nombre": "Eva Martín", "edad": 38, "grupo_edad": "26-40", "codigo_comportamiento": 0},
    {"id_client": 9, "nombre": "Diego Romero", "edad": 24, "grupo_edad": "18-25", "codigo_comportamiento": 1},
    {"id_client": 10, "nombre": "Núria Costa", "edad": 33, "grupo_edad": "26-40", "codigo_comportamiento": 4}
]

comportamientos = {
    0: {"factor": 0.10, "segmento": "Navegación pasiva", "mensaje": "¡Descubre nuestras novedades!"},
    1: {"factor": 0.25, "segmento": "Explorador indeciso", "mensaje": "¿Tienes dudas con los productos?"},
    2: {"factor": 0.50, "segmento": "Checkout incompleto", "mensaje": "¡Estás a un paso! Finaliza la compra."},
    3: {"factor": 0.75, "segmento": "Abandono reciente", "mensaje": "Tu cesta te espera. No te lo pierdas."},
    4: {"factor": 0.90, "segmento": "Abandono crítico", "mensaje": "¡Descuento exclusivo para tu cesta!"}
}

edades = {
    "18-25": 1.20,
    "26-40": 1.00,
    "41-60": 0.85,
    "60+": 0.70
}
```

Funciones Modulares



Pseudocódigo: obtener_segmento_y_factor
* Recorre cada cliente del diccionario.
* Busca en 'comportamientos' usando el código del cliente.
* Asigna el segmento y factor correspondiente al cliente.

```python
"""
    Actualiza el diccionario de clientes añadiendo el segmento,
    el mensaje y el factor de comportamiento basándose en el
    código de comportamiento de cada cliente.
    """

def obtener_segmento_y_factor(diccionario):
    for cliente in diccionario:
        codigo = diccionario[cliente]["codigo_comportamiento"]
        diccionario[cliente]["segmento"] = comportamientos[codigo]["segmento"]
        diccionario[cliente]["factor_comportamiento"] = comportamientos[codigo]["factor"]
```

Pseudocódigo: obtener_factor_edad
* Recorre cada cliente del diccionario.
* Busca en 'edades' usando el grupo_edad del cliente.
* Asigna el factor de edad al cliente.

```python
"""
    Recorre el diccionario de clientes y añade el factor de riesgo
    basado en el grupo de edad de cada uno.
    """

def obtener_factor_edad(diccionario):
    for cliente in diccionario:
        grup = diccionario[cliente]["grupo_edad"]
        diccionario[cliente]["factor_edad"] = edades[grup]
```

Pseudocódigo: calcular_riesgo
* Recorre cada cliente del diccionario.
* Multiplica el factor_comportamiento por el factor_edad.
* Almacena el resultado en 'riesgo_total'.

```python
"""
    Realiza el cálculo del riesgo multiplicando el factor de comportamiento
    por el de edad, redondeando el resultado a 3 decimales.
    """

def calcular_riesgo(diccionario):
    for cliente in diccionario:
        riesgo = diccionario[cliente]["factor_comportamiento"] * diccionario[cliente]["factor_edad"]
        diccionario[cliente]["riesgo_total"] = round(riesgo, 3)
```

**Función: processar_clients**

Pseudocódigo: processar_clients
* Crea un nuevo diccionario organizado por id_client.
* Llama secuencialmente a las funciones de cálculo (obtener datos, factores, riesgo).
* Retorna el diccionario enriquecido.

```python
def processar_clients(clientes_lista):
    diccionario = {
        c["id_client"]: {k: v for k, v in c.items() if k != "id_client"}
        for c in clientes_lista
    }
    obtener_segmento_y_factor(diccionario)
    obtener_factor_edad(diccionario)
    calcular_riesgo(diccionario)
    return diccionario
```

PSEUDOCÓDIGO: mostrar_diccionari
1. Imprime el encabezado y el diccionario procesado.

```python
def mostrar_diccionari(diccionario):
    print("Diccionario de clientes con sus riesgos:\n")
    print(diccionario)
```

Pseudocódigo: mostrar_respuestas
* Crear una lista de clientes a partir de los valores del diccionario.
* Ordenar por riesgo de mayor a menor y mostrar los 3 primeros.
* Filtrar clientes con código 4 para analizar cómo la edad afecta al riesgo.
* Filtrar clientes con riesgo < 0.2 (poco interés).
* Seleccionar el cliente de mayor riesgo (más rentable a recuperar).

```python
def mostrar_respuestas(diccionario):
    clientes_lista = list(diccionario.values())

    print("\n1. Clientes a contactar primero (Retargeting):")
    top_riesgo = sorted(clientes_lista, key=lambda x: x['riesgo_total'], reverse=True)
    for c in top_riesgo[:3]:
        print(f"- {c['nombre']} (Riesgo: {c['riesgo_total']:.2f})")

    print("\n2. Diferencias en clientes con código 4 (Impacto edad):")
    analisis_c4 = [c for c in clientes_lista if c['codigo_comportamiento'] == 4]
    for c in analisis_c4:
        print(f"- {c['nombre']} (Edad: {c['grupo_edad']}, Riesgo: {c['riesgo_total']:.2f})")

    print("\n3. Clientes con bajo interés (Riesgo < 0.2):")
    poc_interes = [c['nombre'] for c in clientes_lista if c['riesgo_total'] < 0.2]
    print(", ".join(poc_interes))

    print("\n4. Cliente más rentable a recuperar:")
    mas_rentable = max(clientes_lista, key=lambda x: x['riesgo_total'])
    print(f"- {mas_rentable['nombre']} (Riesgo: {mas_rentable['riesgo_total']:.2f})")
```

Función: main_marketing

* PSEUDOCÓDIGO: main_marketing
* Orquesta todo el programa llamando a procesar, mostrar diccionario y mostrar respuestas.

```python
"""
    Función orquestadora que gestiona el flujo de procesamiento de los clientes
    y dispara la generación del reporte final.
    """

def main_marketing():
    diccionario = processar_clients(clientes)
    mostrar_diccionari(diccionario)
    mostrar_respuestas(diccionario)
```

```python
# --- EJECUCIÓN ---
if __name__ == "__main__":
    main_marketing()
```

Gestió de risc i priorització de pacients
Treballes com a analista de dades en un centre mèdic especialitzat. Diversos pacients han enviat els seus símptomes a través d'un formulari online, però les dades arriben en un format desordenat i poc estructurat, i l'equip mèdic necessita que els organitzis per poder prioritzar l'ordre d'atenció segons la gravetat.

Objectius
L'objectiu és transformar aquestes dades en informació útil per als metges. A partir dels registres en brut, has de:

* Convertir cada string en una estructura neta (diccionari o similar).

* Separar i normalitzar la llista de símptomes:
minúscules
eliminar espais
dividir correctament per comes
* Assignar un risc a cada símptoma utilitzant el diccionari nivells_risc.

* Calcular el risc total de cada pacient sumant els valors.

* Ordenar els pacients de major a menor risc per determinar la prioritat d'atenció.

Resultat esperat
Un DataFrame o un diccionari amb:

nom
edat
símptomes normalitzats
risc_total
I un apartat final amb els pacients ordenats per prioritat (més risc → primer).

**Algoritmo General:**


Gestión de Riesgo y Priorización de Pacientes

Configuración: Definición de los niveles de riesgo y estructura de datos inicial.

Entrada y Limpieza: Conversión de cadenas crudas a diccionarios estructurados, normalizando síntomas.

Lógica de Negocio: Cálculo de la puntuación total de riesgo sumando los valores asignados a cada síntoma.

Orquestación: Secuencia lógica que encadena la limpieza, el cálculo y la ordenación de los casos.

Salida: Generación de un informe estructurado que muestra a los pacientes ordenados por su nivel de riesgo.

**Desgranaje del Sistema Modular**

```python
# --- DATOS ---
pacientes = [
    "Maria|45|tos, Fiebre Alta, dolor_pecho",
    "Luis|33|dolor_cabeza,faTiga",
    "Sara|67|fiebre alta, dificultad RESPIRAR, tos",
    "Jordi|52|fatiga , tos",
    "Anna|29|fiebre alta,dolor_cabeza"
]

nivells_risc = {
    "fiebre_alta": 3, "dificultad_respirar": 5, "dolor_pecho": 4,
    "tos": 1, "fatiga": 1, "dolor_cabeza": 1
}
```

**Funciones Modulares**

 PSEUDOCÓDIGO: limpiar_datos
1. Recibe la lista de pacientes.
2. Bucle: divide por '|', convierte edad a entero y crea lista limpia de síntomas.
3. Retorna un diccionario con los datos estructurados.

```python
def limpiar_datos(lista_pacientes):
    """
    Recibe la lista de pacientes y devuelve un diccionario con
    la edad y los síntomas normalizados.
    """
    dict_pacientes = {}
    for paciente in lista_pacientes:
        datos = paciente.split("|")
        nombre = datos[0]
        edad = int(datos[1])
        sintomas_sin_normalizar = datos[2].split(',')

        sintomas_normalizados = []
        for sintoma in sintomas_sin_normalizar:
            sintomas_normalizados.append("_".join(sintoma.lower().split()))

        dict_pacientes[nombre] = {
            "edad" : edad,
            "sintomas" : sintomas_normalizados
        }
    return dict_pacientes
```

PSEUDOCÓDIGO: calcular_riesgo
1. Recibe el diccionario de pacientes.
2. Bucle: suma el valor de riesgo de cada síntoma consultando 'nivells_risc'.
3. Guarda el total en la clave 'riesgo'.

```python
def calcular_riesgo(dict_pacientes):
    """
    Calcula el riesgo total de cada paciente según sus síntomas.
    """
    for paciente in dict_pacientes:
        riesgo = 0
        for sintoma in dict_pacientes[paciente]["sintomas"]:
            riesgo += nivells_risc[sintoma]
        dict_pacientes[paciente]["riesgo"] = riesgo
```

 PSEUDOCÓDIGO: ordenar_por_riesgo
 1. Recibe el diccionario con riesgos.
 2. Ordena los elementos basándose en la clave 'riesgo' de forma descendente.
 3. Retorna la lista ordenada.

```python
def ordenar_por_riesgo(dict_pacientes):
    """
    Ordena los pacientes de mayor a menor riesgo.
    """
    return sorted(
        dict_pacientes.items(),
        key=lambda elemento: elemento[1]["riesgo"],
        reverse=True
    )
```

2. Calcula el riesgo.
3. Ordena los resultados y los imprime

**generar_diccionario:**
* Recibe: La lista global clientes.
* Procesa: Reorganiza los datos usando el nombre como clave principal del diccionario.
* Retorna: El nuevo diccionario indexado.

**generar_informe:**

* Recibe: El diccionario de clientes.
* Procesa: Itera sobre el diccionario e imprime filas formateadas con alineación de texto.
* Retorna: Nada (salida en consola).

**generar_analisis:**
* Recibe: El diccionario de clientes.
* Procesa: Utiliza funciones de búsqueda (max) para localizar registros específicos según condiciones de fraude y precio.
* Retorna: Nada (salida en consola).

```python
# Datos de entrada

clientes = [
    {"nom": "Ana", "edat": 42, "codi": 0},
    {"nom": "Carlos", "edat": 30, "codi": 2},
    {"nom": "Isabel", "edat": 55, "codi": 3},
    {"nom": "Jorge", "edat": 40, "codi": 1},
    {"nom": "Marta", "edat": 28, "codi": 0}
]

nivells_seguro = {
    0: {"categoria": "Baix risc", "preu_base": 120, "fraude": 0.01},
    1: {"categoria": "Risc mitjà", "preu_base": 200, "fraude": 0.05},
    2: {"categoria": "Risc alt", "preu_base": 350, "fraude": 0.15},
    3: {"categoria": "Risc crític", "preu_base": 500, "fraude": 0.30}
}
```

**Funciones**

PSEUDOCÓDIGO: añadir_datos
1. Recibe la lista de clientes.
2. Bucle: Busca en 'nivells_seguro' por código y añade datos al diccionario del cliente.
3. No retorna nada (modifica la lista original).

```python
def añadir_datos():
    for cliente in clientes:
        cliente["categoria"] = nivells_seguro[cliente["codi"]]["categoria"]
        cliente["preu_base"] = nivells_seguro[cliente["codi"]]["preu_base"]
        cliente["fraude"] = nivells_seguro[cliente["codi"]]["fraude"]
```

PSEUDOCÓDIGO: calcular_precio_final
1. Recibe la lista de clientes.
2. Bucle: Calcula el precio aplicando la fórmula: preu_base * (1 + edat/100).
3. No retorna nada (añade la clave 'precio_final').

```python
def calcular_precio_final():
    for cliente in clientes:
        cliente["precio_final"] = cliente["preu_base"] * (1 + cliente["edat"]/100)
```

PSEUDOCÓDIGO: generar_diccionario
1. Recibe la lista de clientes.
2. Transforma la lista en un diccionario donde la clave es el nombre.
3. Retorna el nuevo diccionario.

```python
def generar_diccionario():
    return {
        cliente["nom"]: {k: v for k, v in cliente.items() if k != "nom"}
        for cliente in clientes
    }
```

PSEUDOCÓDIGO: generar_informe
1. Recibe el diccionario de clientes.
2. Imprime encabezados y formatea cada fila con alineación fija.

```python
def generar_informe(diccionario):
    linea = "=" * 80
    print(linea)
    print(f"{'INFORME DE CONTROL DE RIESGOS Y PÓLIZAS - COMPAÑÍA ASEGURADORA':^80}")
    print(linea)
    header = f"{'Nombre':<10} | {'Edad':<5} | {'Categoría Riesgo':<18} | {'Precio Base':<12} | {'Precio Final':<12} | {'Fraude':<6}"
    print(header)
    print("-" * len(header))

    for nombre, datos in diccionario.items():
        fraude_porc = f"{int(datos['fraude'] * 100)}%"
        print(f"{nombre:<10} | {datos['edat']:<5} | {datos['categoria']:<18} | ${datos['preu_base']:<11} | ${datos['precio_final']:<11.1f} | {fraude_porc:<6}")
    print(linea)
```

PSEUDOCÓDIGO: generar_analisis
1. Recibe el diccionario.
2. Calcula mediante funciones 'max' los valores críticos.
3. Imprime el resumen de auditoría.

```python
def generar_analisis(diccionario):
    mas_fraude = max(diccionario.items(), key=lambda x: x[1]['fraude'])
    mas_caro = max(diccionario.items(), key=lambda x: x[1]['precio_final'])

    print("\n ANÁLISIS Y AUDITORÍA")
    print(f"» Cliente con más probabilidad de fraude:\n-> {mas_fraude[0]} ({int(mas_fraude[1]['fraude']*100)}% de probabilidad).")
    print(f"\n» Cliente con el precio final más alto:\n-> {mas_caro[0]} (${mas_caro[1]['precio_final']:.1f} debido a su edad de {mas_caro[1]['edat']} años).")
    print(f"\n» Prioritario revisar manualmente:\n-> {mas_fraude[0]} (Posee una categoría de '{mas_fraude[1]['categoria']}' con un {int(mas_fraude[1]['fraude']*100)}% de riesgo de fraude).")
```

PSEUDOCÓDIGO: main
1. Llama a las funciones de procesamiento.
2. Genera los informes y el análisis final.

```python
def main():
    añadir_datos()
    calcular_precio_final()
    diccionario = generar_diccionario()

    generar_informe(diccionario)
    generar_analisis(diccionario)

if __name__ == "__main__":
    main()
```
